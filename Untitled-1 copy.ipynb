```python
import tkinter as tk
from tkinter import scrolledtext
import sys
import minecraft_launcher_lib
import sys
import subprocess
import os
import requests
from pathlib import Path

def imprimir_algo():
        usuario = "pepito"
        print(f"Ingresando al juego como: {usuario}")
        # Actualizar la variable con el texto ingresado en el cuadro de texto
        vanilla_version = "1.20.1"
        minecraft_directory = minecraft_launcher_lib.utils.get_minecraft_directory()
        current_max = 0
        print(minecraft_directory)

        def set_status(status: str):
            print(status)

        def set_progress(progress: int):
            global current_max
            if current_max != 0:
                percentage = (progress / current_max) * 100
                print(f"({percentage:.0f}%)")

        def set_max(new_max: int):
            global current_max 
            current_max = new_max


        installed_versions = minecraft_launcher_lib.utils.get_installed_versions(minecraft_directory)
        print(installed_versions)

        callback = {
                "setStatus": set_status,
                "setMax": set_max,
                "setProgress": set_progress,
        
            }

        target_id = f'fabric-loader-0.15.3-{vanilla_version}'

            # Verificar si el diccionario con 'id': 'fabric-loader-0.15.3-1.20' está en la lista
        is_present = any(item['id'] == target_id for item in installed_versions)

        if is_present:
            print(f"Ejecutando...")

        else:
            print(f"Instalando... (puede tomar unos minutos)")
            minecraft_launcher_lib.fabric.install_fabric(vanilla_version, minecraft_directory, callback=callback)

                # Crea la carpeta 'mods' si no existe
        print("verificando mods...")
        mods_directory = Path(minecraft_directory) / "mods"
        mods_directory.mkdir(parents=True, exist_ok=True)

        mods_to_install = {
                "Cobblemon": "https://cdn.modrinth.com/data/MdwFAVRL/versions/uWAkNUxZ/Cobblemon-fabric-1.4.1%2B1.20.1.jar",
                "Fabric API": "https://cdn.modrinth.com/data/P7dR8mSH/versions/YblXfKtI/fabric-api-0.91.0%2B1.20.1.jar",
                "tradersNPC":"https://mediafilez.forgecdn.net/files/4769/890/taterzens-fabric-1.20.1-1.11.7.jar",
                "addonTraderNPC":"https://mediafilez.forgecdn.net/files/3825/962/tradernpcs-1.2.0-fabric.jar",
                "terraform":"https://cdn.modrinth.com/data/8oi3bsk5/versions/FgvUosFH/Terralith_1.20.4_v2.4.11.jar",
                "cobblepedia":"https://cdn.modrinth.com/data/2obPz7jf/versions/dYSp3ZsV/Cobblepedia-Fabric-0.5.2.jar",
                "xaerominimap":"https://cdn.modrinth.com/data/1bokaNcj/versions/KSDcShTK/Xaeros_Minimap_23.9.3_Fabric_1.20.jar",
                "waystones":"https://cdn.modrinth.com/data/LOpKHB2A/versions/36jYwMdH/waystones-fabric-1.20-14.1.3.jar",
                "Balm":"https://cdn.modrinth.com/data/MBAkmtvl/versions/qIAY6mKt/balm-fabric-1.20.1-7.2.1.jar",
                "Compass":"https://cdn.modrinth.com/data/fPetb5Kh/versions/NovIXDxY/NaturesCompass-1.20.1-2.2.3-fabric.jar",
                "patchouli":"https://cdn.modrinth.com/data/nU0bVIaL/versions/Y6tuH1cn/Patchouli-1.20.1-84-FABRIC.jar",
                "spawnNotifi":"https://cdn.modrinth.com/data/LPuJjiQz/versions/s6xwimto/cobblemon-spawn-notification-1.4-fabric-1.2.0.jar",
                "kotlin":"https://cdn.modrinth.com/data/Ha28R6CL/versions/JjrWZ7m8/fabric-language-kotlin-1.10.17%2Bkotlin.1.9.22.jar",
                "mountable":"https://cdn.modrinth.com/data/Cj8jrU7n/versions/FB77FTsq/cobblemounts-1.2.1.jar",
                "catchEXP":"https://cdn.modrinth.com/data/LBl4Qguc/versions/EHJteu1u/cobblemon-capturexp-1.4-fabric-1.0.2.jar",
                "gooeylib":"https://cdn.modrinth.com/data/KzjvlKEN/versions/BXIm53Zb/GooeyLibs-Fabric-3.0.0%2B1.20.1.jar",
                "impactor":"https://cdn.modrinth.com/data/LdBYVaPS/versions/KzE0nIJR/Impactor-Fabric-5.1.1%2B1.20.1.jar",
                "gts":"https://cdn.modrinth.com/data/UTjx5ypK/versions/bsTWNajU/Gts-1.1.3-fabric.jar",
                "hunt":"https://cdn.modrinth.com/data/QMAwqHhY/versions/GdXug8ZS/Hunt-1.0.7-fabric.jar"





            }

        installed_mods = set()

        for mod_name, mod_url in mods_to_install.items():
            mod_path = mods_directory / f"{mod_name}.jar"

            # Verificar si el mod ya está instalado
            if mod_path.exists():
                print(f"{mod_name} ya está instalado.")
                installed_mods.add(mod_name)
            else:
                # Descargar e instalar el mod
                response = requests.get(mod_url)
                with open(mod_path, 'wb') as mod_file:
                    mod_file.write(response.content)
                print(f"{mod_name} instalado correctamente.")

            # Verificar si todos los mods están instalados
        if len(installed_mods) == len(mods_to_install):
            print("Todos los mods están instalados. Mods cargados.")


        options = {
            "username": usuario,
            "uuid": "",
            "token": ""
            }
        import time
        minecraft_command = minecraft_launcher_lib.command.get_minecraft_command(target_id, minecraft_directory, options)
        subprocess.run(minecraft_command)

imprimir_algo()
```
